```python
#7
from pyspark.sql.functions import count
df_races.filter(df_races.ano == 2010).agg(count('id_corrida').alias("Total de corridas no ano de 2010")).show()
```

```python
#8
from pyspark.sql.functions import sum
df_results.join(df_drivers, "id_piloto").groupBy("prim_nome", "ult_nome", "dt_nasc", "nacionalidade").agg(sum("posicao").alias("total de pontos")).show()
```

```python
#9
df_results.join(df_drivers, "id_piloto").filter(df_results.posicao == 1).groupBy("prim_nome", "ult_nome", "dt_nasc", "nacionalidade").count().withColumnRenamed("count", "total de vitÃ³rias").show()
```

```python
#10
df_results.join(df_drivers, "id_piloto").join(df_races, "id_corrida").filter((df_drivers.prim_nome == "Michael") & (df_drivers.ult_nome == "Schumacher") & (df_results.posicao == 1)).select("prim_nome", "ult_nome", "etapa", "ano").show()
```

```python
#11
df_results.join(df_drivers, "id_piloto").groupBy("nacionalidade").agg(sum("posicao").alias("total de pontos")).orderBy('total de pontos', ascending=False).show()
```
